```python
import pandas as pd

# Load the CSV data to understand its structure
file_path = '../data/rmsdByRes_ref_gzt.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
data.head()
```

```python
import matplotlib.pyplot as plt

# Plotting RMSD vs. Residue Index
plt.figure(figsize=(10, 6))
plt.plot(data['residueId'], data['allAtomRMSD'], linestyle='-', color='blue')
plt.title('RMSD vs. Residue Index')
plt.xlabel('Residue Index')
plt.ylabel('RMSD (Å)')
plt.grid(True)
plt.show()
```
